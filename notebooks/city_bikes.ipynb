# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
import os
import numpy as np

url = "https://api.citybik.es/v2/networks" ##Url to get all networks
response = requests.get(url) ##get response
data = response.json() #establish connection
networks = pd.DataFrame(data['networks']) #create dataframe with all Networks to determine which ones are available
print(networks['id'].unique()) #print available networks
'''
Importing the API key from the environment variable

The API key is used to access the CityBikes API. It is stored in the environment variables.
'''
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# Making a request to City Bikes for Vancouver, BC
'''
Making a request to citybikes API to get location information from Mobi Bikes
'''
network_id = "mobibikes" ##Name of Bikeshare program in Vancouver BC
url = f"https://api.citybik.es/v2/networks/{network_id}" #url to make request from 
response = requests.get(url)
data = response.json()
```

Put your parsed results into a DataFrame.

```python
# Load station data into DataFrame
stations_df = pd.DataFrame(data['network']['stations'])

# Display relevant information
print(stations_df[['name', 'free_bikes', 'empty_slots', 'latitude', 'longitude']].head(25)) #Testing to ensure that it works 

#Add a new column to stations_df. Adding together free_bikes and empty_slots to a new column called total_bikes to determine total bikes each station can hold to account for variation in time of day request is pulled


stations_df['total_bikes'] = stations_df['free_bikes'] + stations_df['empty_slots']
stations_df

##Save stations_df as CSV

stations_df.to_csv('stations_df.csv', index=False)
```

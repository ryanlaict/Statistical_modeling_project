```python
#Importing various packages to install
import requests
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

from sklearn import linear_model, datasets
```

Build a regression model.

```python
#Setting up the Dataset for regression model. 
data = pd.read_csv('combined_dataset_cleaned.csv') #Making a copy of the dataset

data
```

```python
'''Looking at dependent variable of Price, using backward regressions to build the most reliable output. Removing Highest P Values until the model is sufficent'''

regression_df = data[['Price_y','Rating_y','Distance','review_count','empty_slots','total_bikes']]

y = regression_df['review_count']
X = regression_df.drop(columns=['review_count','empty_slots']) 

X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Display the summary of the regression results
print(model.summary())
```

Provide model output and an interpretation of the results. 

The regression model provided several key statistics. The first is that this model is able to explain roughly 25% of the variability in Review counts and that with a F value of 32, it shows that the results of this table are meaningfully related to Review Counts in this dataset. 

Price showed the greatest relationshp with Review counts, showing that with an increase in Price, it created roughly 118 more reviews for a location. Ratings also showed a strong relationship with an increase in Rating causing roughly 54 more reviews. Both these valus also have statistically significant p-values(0.000), meaning that the probability of these results being random are null. 

The size of bikes that the nearest station could hold was also a factor. A positive relationship between the variables indicates that an increase in the total number of bikes correlates with a small increase in review counts. An increase in station size was shown to have an increase of 6 reviews for a place. 

Distance had a weak relationship, both with its affect on review counts but with a P-value of above 0.05, it means that we can not definitely confirm that any positive affect on review counts were not simply due to chance. 

The results of this model showed that Price and Ratings had the biggest impact on the amount of reviews that a place received. The size of a bike station was third in its impact and the distance from a bike station however was not as impactful. From a business perspective based on this model, the focus would be to improve a restaurants overall ratings and price in order to generate more reviews for a place. 

There is a key of multicolinarity, with Ratings of a place being closely related to its review count and it affecting each other and further analysis will need to be done to confirm the impact that they have on each other and if they should be taken into separate consideration.
